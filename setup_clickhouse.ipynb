```python
# encoding:UTF-8
import pandas as pd
import clickhouse_connect

ch_client = clickhouse_connect.get_client(host='localhost', username='default', database='finance')

try:
    result = ch_client.query('SELECT * FROM bars_daily')
    df = pd.DataFrame(dict(zip(result.column_names, result.result_columns)))
finally:
    ch_client.close()
df
```
