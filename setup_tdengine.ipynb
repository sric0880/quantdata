### Write

将csv文件复制到容器中，并执行`finance.tdengine.sql`

注意：数据库默认是系统时区，不设置的话，默认是北京时区，需要手动修改容器的配置，并从容器内部执行导入。从外部导入通过taos命令行导入，还是会变成北京时区。

### Read

```python
# encoding:UTF-8
import taos
import pandas as pd

tdengine = taos.connect(
    host='localhost',
    port=6030,
    user="root",
    password="taosdata",
    database="finance",
    timezone="UTC",
)

try:
    result = tdengine.query("SELECT * FROM bars_daily_000001_sz;")
    df = pd.DataFrame(result.fetch_all(), columns=[field.name for field in result.fields])
finally:
    tdengine.close()
df
```
